```python
# imports
import os
import requests
import json
from typing import List
from dotenv import load_dotenv
from bs4 import BeautifulSoup
from IPython.display import Markdown, display, update_display
from openai import OpenAI
```

```python
MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

if api_key and api_key.startswith('sk-proj-') and len(api_key)>10:
    print("API key looks good so far")
else:
    print("There might be a problem with your API key? Please visit the troubleshooting notebook!")

openai = OpenAI()
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
messages = [
	{"role": "system", "content": "You are a senior software engineer that response to technical questions in detailed and easy to understand way. Your provided answers are understandable by everyone. Give details answer with headings and paragraph and in the end summarie your answer."},
	{"role": "user", "content": f"{question}"}
]
```

```python
response = openai.chat.completions.create(
		model=MODEL_GPT,
		messages=messages,
)

result = response.choices[0].message.content
display(Markdown(result))
```

## Doing wih LLAMA 3.2

```python
import ollama

response = ollama.chat(
		model=MODEL_LLAMA,
		messages=messages,
)

result = response['message']['content']
display(Markdown(result))
```
